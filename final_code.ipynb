```python
import wikipediaapi
import requests
import pandas as pd
from bs4 import BeautifulSoup

#installing scraping and visualization
```

```python
wikiurl="https://en.wikipedia.org/wiki/List_of_sovereign_states_by_tax_revenue_to_GDP_ratio"
table_class="wikitable sortable jquery-tablesorter"
response=requests.get(wikiurl)
print(response.status_code)

#check if we are allowed to scrape wikipedia
#200 means yea
```

```python
soup = BeautifulSoup(response.text,'html.parser')
countrytable=soup.find('table',{'class':"wikitable"})
print(countrytable)

#prints table in html format
```

```python
userdata = pd.read_html(str(countrytable))
userdata = pd.DataFrame(userdata[0])
userdata

#userdata dataframe has all the information from the table
```

```python
userdata.columns

#inspect the columns
```

```python
len(userdata.columns)

#find the length of the columns
```

```python
userdata.columns = ["Continent", "Country", "Tax Revenue (% of GDP)", "GDP ($ billion)", "Tax Revenue ($ billion)", "Government Expenditure (% of GDP)", "Public debt (% of GDP)"]

#gave appropriate names to the columns
```

```python
userdata = userdata.iloc[0:181]
#using iloc function here because 181 - 185 have NA in their values
#having NA in their values messes it up so they were spliced
#limitation
userdata
```

```python
summarydata = userdata[["Continent", "Country","Tax Revenue (% of GDP)","Government Expenditure (% of GDP)"]].copy()
summarydata["Tax Revenue (% of GDP)"] = pd.to_numeric(summarydata["Tax Revenue (% of GDP)"],errors = 'coerce')
summarydata["Government Expenditure (% of GDP)"] = pd.to_numeric(summarydata["Government Expenditure (% of GDP)"],errors = 'coerce')
summarydata
```

```python
processeddata = summarydata[["Tax Revenue (% of GDP)","Government Expenditure (% of GDP)"]].copy()
processeddata
#to do k means, we need only numbers
```

```python
from sklearn.cluster import KMeans
kmeans_model = KMeans(n_clusters=3, random_state=0, n_init=1)
kmeans_model.fit(processeddata)

#clusters show how we're gonna segment the data
#low economic health, medium, high
```

```python
list_of_performance = list(kmeans_model.labels_)
list_of_performance

#how health it is with balancing
```

```python
summarydata["Labels"] = list_of_performance
summarydata

#adds the list to the table
#ex: saudi arabia doesn't tax and has a low score
#not accurate
```

```python
continent_dataframes = {}
for continent, group in summarydata.groupby('Continent'):
    continent_dataframes[continent] = group
```

```python
asia_data = continent_dataframes.get('Asia')
asia_data.value_counts("Labels")

#visualize all of the regions' tables
#bar graph
```
